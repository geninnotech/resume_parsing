```python
from dotenv import load_dotenv
import os

load_dotenv(override=True)

API_KEY = os.environ.get("GROQ_API_KEY")
MODEL_NAME=os.environ.get("MODEL_NAME")        
```

```python
from prompts.prompt_builder import make_ocr_prompt, make_cv_prompt
from llm.llm_modules import groqLLM
```

```python
class resume_parsing:
    def __init__(self):
        self.llm = groqLLM(api_key=API_KEY, model=MODEL_NAME)
        
    def ocr(self, pdf_path):
        ocr_prompt = make_ocr_prompt()
        ocr_text = self.llm.infer(prompt=ocr_prompt, pdfs=[pdf_path])
        return ocr_text
    
    def get_structured_cv(self, resume_pdf):
        ocr_text = self.ocr(resume_pdf)
        json_prompt = make_cv_prompt(ocr_text)
        structured_json = self.llm.infer(prompt=json_prompt)
        return structured_json
    
    def bulk_structured_cv(self, pdfs, output_folder):
        """
        pdfs can be a list of paths or pdf folder
        """
        if not os.path.exists(output_folder):
            os.makedirs(output_folder)

        if isinstance(pdfs, str):
            if os.path.exists(pdfs):
                pdfs = [os.path.join(pdfs, i) for i in os.listdir(pdfs) if i.endswith(".pdf")]

        elif isinstance(pdfs, list):
            pdfs = [i for i in pdfs if os.path.exists(i) and i.endswith(".pdf")]

        for n, pdf in enumerate(pdfs):
            print(f"Processing CV {n+1}: {pdf}")
            structured_json = self.get_structured_cv(pdf)
            with open(os.path.join(output_folder, os.path.basename(pdf) + ".json"), "w") as f:
                f.write(structured_json)
            
            print(f"Structured CV {n+1} saved to {os.path.basename(pdf) + '.json'}")


```

```python
parser = resume_parsing()
```
